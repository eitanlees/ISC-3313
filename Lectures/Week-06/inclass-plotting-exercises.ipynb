```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exercise 1

```python
def my_function(x, a, b):
    return np.cos(a * x) * np.exp(-b * x)
```

```python
x = np.linspace(0, 5)
```

```python
plt.plot(x, my_function(x, 2, 0.5))
plt.plot(x, my_function(x, 3, 0.7));
```

## Exercise 2

```python
plt.plot(x, np.sin(x), '--', color='gold')
plt.plot(x, np.sin(x - 1), '-', color='seagreen')
plt.plot(x, np.sin(x - 2), ':', color='crimson');
```

## Exersice 3

```python
V = lambda r, sigma: 4 * ((sigma / r) ** 12 - (sigma / r) ** 6)
```

```python
r = np.linspace(0.95, 3)
```

```python
plt.plot(r, V(r, 1), '--', label = "Example One")
plt.plot(r, V(r, 1.5), ':', label = "Example Two")

plt.ylim(-2, 2)
plt.xlabel("r")
plt.ylabel("V(r)")
plt.title("Lennard Jones Potential")
plt.legend();
```
