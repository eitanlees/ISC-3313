```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exercise One

```python
xn = np.array([0.        , 0.62831853, 
              1.25663706, 1.88495559, 
              2.51327412, 3.14159265, 
              3.76991118, 4.39822972, 
              5.02654825, 5.65486678,
              6.28318531])

yn = np.array([0.            ,  5.87785252e-01,  
              9.51056516e-01,  9.51056516e-01, 
              5.87785252e-01,  1.22464680e-16, 
              -5.87785252e-01, -9.51056516e-01,
              -9.51056516e-01, -5.87785252e-01, 
              -2.44929360e-16])
```

```python
from scipy.interpolate import interp1d
```

```python
linear_results = interp1d(xn, yn, kind='linear')
quadratic_results = interp1d(xn, yn, kind='quadratic')

x = np.linspace(xn.min(), xn.max(), 100)
plt.plot(xn, yn, 'o', label='Data')
plt.plot(x, linear_results(x), label='Linear')
plt.plot(x, quadratic_results(x), label='Quadratic')
plt.legend()
```

## Exercise Two

```python
from scipy.stats import gamma
```

```python
dist = gamma(2)
r = dist.rvs(1000, random_state=0)
```

```python
x = np.linspace(0, 10, 200)
plt.hist(r, bins=20, density=True)
plt.plot(x, dist.pdf(x), linewidth=3);
```

```python
params = gamma.fit(r)
print(f'The shape parameter was found to be {params[0]:.4f}')
```
