```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.array([0,10,20,30,40])
x = np.array([0,50,100,120,125])

dt = t[1] - t[0]
```

```python
v_start = (x[1] - x[0])/(dt)
v_middle = (x[2:] - x[:-2]) / (2 * dt)
v_stop = (x[-1] - x[-2])/(dt) 

v = np.concatenate(([v_start], v_middle, [v_stop]))
```

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(t, x, 'o-')
ax[0].set(ylabel='Position', xlabel='Time')

ax[1].plot(t, v, '-o')
ax[1].set(ylabel='Velocity', xlabel='Time')

plt.tight_layout()
```

```python
f = lambda x: np.sin(x)**2
true_I = 5 - np.sin(20)/4

N_list = [25, 50, 100]
a = 0
b = 10

Error = []

for N in N_list:
    h = (b - a)/N
    x = np.linspace(a, b, N + 1)
    I = np.sum(f(x[:-1]))*h
    Error.append(abs(I - true_I))
```

```python
plt.plot(N_list, Error, 'o-')
```

```python
def right_riemann(f, N, a, b):
    h = (b - a)/N
    x = np.linspace(a, b, N + 1)
    I = np.sum(f(x[:-1]))*h
    return I
```

```python
right_riemann(f, 10000, 0, 10)
```
